```python
# init lematizer / stop / stemmer
import nltk
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()
stop = stopwords.words('english')
stop.extend(stops_words)
stemmer = SnowballStemmer("english")
```

```python
def nltk_pos_tagger(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:          
        return None

def lemmatize_sentence(sentence):

    nltk_tagged = nltk.pos_tag(nltk.word_tokenize(sentence))  
    wordnet_tagged = map(lambda x: (x[0], nltk_pos_tagger(x[1])), nltk_tagged)
    lemmatized_sentence = []
    
    for word, tag in wordnet_tagged:
        if tag is None:
            lemmatized_sentence.append(word)
        else:        
            lemmatized_sentence.append(lemmatizer.lemmatize(word, tag))
    return " ".join(lemmatized_sentence)
```

```python
mydata['without_stopwords'] = mydata['plot_synopsis']
# enlever les majuscules
mydata['without_stopwords'] = mydata['without_stopwords'].str.lower()
mydata['without_stopwords'] = mydata['without_stopwords'].str.replace(r'[\W\s]+', ' ')
# enlever les nombres
mydata['without_stopwords'] = mydata['without_stopwords'].str.replace(r'\d', ' ')
# remove stop words
mydata['without_stopwords'] = mydata['without_stopwords'].swifter.apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

# lemmantize les 100 premiers films (donc notre dataset)
for i in range (0,nb_film) :
  mydata['without_stopwords'][i] = lemmatize_sentence(mydata['without_stopwords'][i])
  
print(mydata['without_stopwords'])
```

```python
from nltk.tokenize import word_tokenize
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

def best_result(id_film, coll):
    film_without_stop = mydata.loc[id_film][coll]
    tokenized = word_tokenize(film_without_stop)
    tokenized.sort()
    # print(tokenized)
    
    best_index = -1
    max_compare = 0
    
    for index in range(0,nb_film):
        # get film
        if index != id_film:
            film_comp_without_stop = mydata.loc[index][coll]
            tokenized_comp = word_tokenize(film_comp_without_stop)
            tokenized_comp.sort()
            compared = 0    
            last_index = 0
            for tok in tokenized:
                for i in range(last_index, len(tokenized_comp)):
                    tok_comp = tokenized_comp[i]
                    if tok > tok_comp:
                        pass
                    elif tok == tok_comp:
                        compared+=1
                        last_index = i
                        break
                    else :
                        last_index = i
                        break
            if max_compare<compared:
                max_compare = compared
                best_index = index
        
    print("best value {} eg \n \033[1;32m {} \033[1;0m->\033[1;10m {}".format(max_compare, mydata.loc[id_film]['title'], mydata.loc[best_index]['title']))
    print("id: {} -> {}".format(id_film, best_index))

best_result(1, 'without_stopwords')
```

```python
import nltk
from nltk.corpus import wordnet
import pprint

syno_saved = []

def replace_synonyms(sentence):
    words = nltk.word_tokenize(sentence)
    new_words = []
    for word in words:
        synonyms = wordnet.synsets(word)
        if synonyms:
            common_syn = synonyms[0]
            biggest_num = 0
            syno = ""
            for syn in synonyms:
                if syn.lemmas()[0].name() in syno_saved:
                    syno = syn.lemmas()[0].name()
                    break
                num = str(syn.lemmas()[0]).split('.')[2]
                try:
                    if int(num) > biggest_num:
                        biggest_num = int(num)
                        syno = syn.lemmas()[0].name()
                except:
                    pass
            if syno not in syno_saved:
                syno_saved.append(syno)
            new_words.append(syno)
        else:
            new_words.append(word)
    return ' '.join(new_words)

# test de la fonction
print(replace_synonyms("Whereof one cannot speak thereof one must been silent"))
print(replace_synonyms("Whereof one cannot talk thereof one must be silent"))
```

```python
mydata['with_syno'] = mydata['without_stopwords']
mydata['with_syno'] = mydata['with_syno'].swifter.apply(replace_synonyms)
```
